```python
import pandas as pd
df=pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import datetime as dt
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df.info()
```

```python
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year
df.head()
```

```python
#MONTHLY SALES
import matplotlib.pyplot as plt
import seaborn as sns
df_monthly=df.groupby('Order Month')['Sales'].sum().to_frame()
df_monthly
```

```python
#VISUALIZE
plt.figure(figsize=(10,5))
plt.plot(df_monthly['Sales'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
#SALES BY CATEGORY
df_category=df.groupby('Category')['Sales'].sum().to_frame()
df_category

#VISUALIZE
sns.barplot(x=df_category.index, y='Sales', data=df_category)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
plt.pie(df_category['Sales'], labels=df_category.index, autopct='%1.1f%%')
plt.title('Sales by Category')
plt.show()
```

```python
#SALES BY SUB-CATEGORY
df_sub_category=df.groupby('Sub-Category')['Sales'].sum().to_frame()
df_sub_category
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=df_sub_category.index, y='Sales', data=df_sub_category)
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.title('Sales by Sub-Category')
plt.show()
```

```python
#PROFIT ANALYSIS
df_monthly_profit=df.groupby('Order Month')['Profit'].sum().to_frame()
df_monthly_profit

#VIZUALISE
plt.figure(figsize=(10,5))
plt.plot(df_monthly_profit['Profit'])
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Monthly Profit')
plt.show()
```

```python
#CATEGORY WISE PROFIT
df_category_profit=df.groupby('Category')['Profit'].sum().to_frame()
df_category_profit

#VISUALIZE
sns.barplot(x=df_category_profit.index, y='Profit', data=df_category_profit)
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Category Wise Profit')
plt.show()
```

```python
#SUB-CATEGORY WISE PROFIT
df_sub_category_profit=df.groupby('Sub-Category')['Profit'].sum().to_frame()
df_sub_category_profit

#VISUALIZE
plt.figure(figsize=(20,5))
sns.barplot(x=df_sub_category_profit.index, y='Profit', data=df_sub_category_profit)
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.title('Sub-Category Wise Profit')
plt.show()
```

```python
#SALES AND PROFIT ANALYSIS BY CUSTOMER SEGMENT
df_customer_sales_profit=df.groupby('Segment')[['Sales','Profit']].sum()
df_customer_sales_profit
```

```python
#VIZUALISE BOTH SALES VS PROFIT
df_melted = df_customer_sales_profit.reset_index().melt('Segment', var_name='Metric', value_name='Amount')
print(df_melted)
plt.figure(figsize=(10,5))
sns.barplot(x='Segment', y='Amount', hue='Metric', data=df_melted)
plt.xlabel('Customer Segment')
plt.ylabel('Sales/Profit')
plt.title('Sales and Profit by Customer Segment')
plt.show()
```

```python
#SALES TO PROFIT RATIO
df_customer_sales_profit['Sales to Profit Ratio'] = df_customer_sales_profit['Sales'] / df_customer_sales_profit['Profit']
df_customer_sales_profit
```
